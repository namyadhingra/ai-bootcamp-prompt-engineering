```python
import getpass
import os

# Set up your OpenAI client
if not os.getenv("OPENAI_API_KEY"):
    secret_key = getpass.getpass("Enter your OpenAI API key: ")
    os.environ["OPENAI_API_KEY"] = secret_key
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts.chat import ChatPromptTemplate

#from_messages-function-takes a list of tuples
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that generates company names"),
    ("human", "{text}")
])
#every time you have curly braces (like {text} up here)
#you will have to add them to the dictionary, which will then invoke the prompt template


result = chat_prompt_template.invoke({
    "text": "What would be a good company name for a company that makes colorful socks?"
})

print(result)
print(type(result))
```

```python
# or we can go:

model = ChatOpenAI(model="gpt-4o-mini")
ai_llm_result = model.invoke(result)
print(ai_llm_result.content)
```

##Streaming

Concept that allows you to get the output of an LLM token by token rather than waiting for the entire response.

```python
%pip install langchain langchain_openai langchain_community --upgrade --quiet
#API key taking from above

from langchain import ChatOpenAI
from langchain_core.messages import HumanMessages
```

```python
model = ChatOpenAI()
model.stream("What is the capital of the moon?")
# using .stream instead of .invoke
for chunk in model.stream("What is the capital of the moon?"):
    print(chunk.content)

#flush the input stream
import sys
for chunk in model.stream("What is the capital of the moon?"):
    print(chunk.content, end='', flush=True)
    sys.stdout.flush()
```

```python
for chunk in model.stream([HumanMessage(content="What is the capital of the moon?")]):
  print(chunk.content, end='', flush=True)
  sys.stdout.flush()

```
